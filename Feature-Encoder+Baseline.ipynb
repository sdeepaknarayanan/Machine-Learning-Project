## A Baseline Formulation

```python
sample_size     = 100     # latent sample size (i.e., 100 random numbers)
g_hidden_size   = 128
d_hidden_size   = 128
leaky_alpha     = 0.01
g_learning_rate = 0.001  # learning rate for the generator
d_learning_rate = 0.001   # learning rate for the discriminator
epochs          = 10
batch_size      = 64      # train batch size
eval_size       = 16      # evaluate size
smooth          = 0.01
```

```python
def preprocess(x):    
    x = x.reshape(-1, 784) # 784=28*28
    x = np.float64(x)
    x = (x / 255 - 0.5) * 2
    x = np.clip(x, -1, 1)
    return x
```

```python
def deprocess(x):
    x = (x / 2 + 1) * 255
    x = np.clip(x, 0, 255)
    x = np.uint8(x)
    x = x.reshape(28, 28)
    return x
```

```python
def make_trainable(model, trainable):
    for layer in model.layers:
        layer.trainable = trainable
```

```python
def make_labels(size):
    return np.ones([size, 1]), np.zeros([size,1])
```

```python
# X_train_real = preprocess(trainmnist)
# X_test_real  = preprocess(testmnist)
```

```python
y_train_real, y_train_fake = make_labels(batch_size)
y_eval_real,  y_eval_fake  = make_labels(eval_size)
```

```python
generator = Sequential()
generator.add(Conv2DTranspose(256, (4,4), input_shape=(128,1,1), strides = (1,1),data_format="channels_first", padding="same"))
generator.add(Activation('relu'))
generator.add(Conv2DTranspose(128, (4,4), strides = (2,2),data_format="channels_first", padding="same"))
generator.add(BatchNormalization())
generator.add(Activation('relu'))
generator.add(Conv2DTranspose(64, (4,4),strides = (4,4),data_format="channels_first", padding="same"))
generator.add(BatchNormalization())
generator.add(Activation('relu'))
generator.add(Conv2DTranspose(1, (4,4), strides = (4,4),data_format="channels_first", padding="same"))
generator.add(Activation('tanh'))


generator.summary()
```

```python
discriminator = Sequential()
discriminator.add(Conv2D(64, (4,4), input_shape=(1,32,32,), strides = (2,2),data_format="channels_first", padding="same"))
discriminator.add(Activation('relu'))
discriminator.add(Conv2D(128, (4,4), strides = (2,2),data_format="channels_first", padding="same"))
discriminator.add(BatchNormalization())
discriminator.add(Activation('relu'))
discriminator.add(Conv2D(256, (4,4),strides = (2,2),data_format="channels_first", padding="same"))
discriminator.add(BatchNormalization())
discriminator.add(Activation('relu'))
discriminator.add(Conv2D(128, (4,4), strides = (2,2),data_format="channels_first", padding="same"))
discriminator.add(Activation('tanh'))
discriminator.add(Conv2D(1, (1,1), strides = (2,2),data_format="channels_first", padding="same"))
discriminator.add(Flatten())
discriminator.add(Dense(1))
discriminator.add(Activation('sigmoid'))


discriminator.summary()
```

```python
gan = Sequential([
    generator,
    discriminator
])

gan.summary()
```

```python
generator.compile(loss='categorical_crossentropy', optimizer = 'adadelta', metrics = ['accuracy'])
discriminator.compile(optimizer=Adam(lr=d_learning_rate), loss='binary_crossentropy')
gan.compile(optimizer=Adam(lr=g_learning_rate), loss='binary_crossentropy')
```

```python
iter_ = 0
```

```python
X_test_mnist.shape
```

```python
losses = []

for e in range(epochs):
    for i in range(len(trainmnist)//batch_size):
        X_batch_real = trainmnist[i*batch_size:(i+1)*batch_size]
        feature_encoder = new_model.predict(x_train[i*batch_size:(i+1)*batch_size])
        X_batch_fake = generator.predict_on_batch(feature_encoder)
        iter_+=1
        #print(X_batch_real.shape)
#         try:
#         print(len(X_batch_fake),len(y_train_fake))
        tmp = X_batch_real.reshape((64,1,32,32))
        make_trainable(discriminator, True)
        discriminator.train_on_batch(tmp, y_train_real * (1 - smooth))
        discriminator.train_on_batch(X_batch_fake, y_train_fake)
    
        make_trainable(discriminator, False)
        gan.train_on_batch(feature_encoder, y_train_real)
  
  
    X_eval_real = testmnist[np.random.choice(len(X_test_mnist), eval_size, replace=False)]
    tmp2 = X_eval_real.reshape((eval_size,1,32,32))
    feature_encoder = new_model.predict(x_train[:eval_size])
    X_eval_fake = generator.predict_on_batch(feature_encoder)
  
    d_loss  = discriminator.test_on_batch(tmp2, y_eval_real)
    d_loss += discriminator.test_on_batch(X_eval_fake, y_eval_fake)
    g_loss  = gan.test_on_batch(feature_encoder, y_eval_real) # we want the fake to be realistic!
    
    losses.append((d_loss, g_loss))
    print("Epoch: {:>3}/{} Discriminator Loss: {:>6.4f} Generator Loss: {:>6.4f}".format(e+1, epochs, d_loss, g_loss))
```

```python
losses = np.array(losses)

fig, ax = plt.subplots()
plt.plot(losses.T[0], label='Discriminator')
plt.plot(losses.T[1], label='Generator')
plt.title("Training Losses")
plt.legend()
plt.show()
```
