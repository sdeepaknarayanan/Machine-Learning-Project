## Dataset Shape

```python
print("MNIST TRAN SET:  ", list(mnist_trainset.train_data.shape))
print("MNIST TEST SET:  ", list(mnist_testset.test_data.shape))
print("SVHN TRAIN SET:  ", list(svhn_trainset.data.shape))
print("SVHN TEST SET:   ", list(svhn_testset.data.shape))
print("SVHN EXTRA SET:  ", list(svhn_extraset.data.shape))
```

## Dataset Label Distribution

```python
plt.hist(list(mnist_trainset.train_labels),color='r',label="TRAIN SET")
plt.hist(list(mnist_testset.test_labels),color='b',label="TEST SET")
plt.title("Plot of MNIST DATASET")
plt.xlabel("Data Labels")
plt.ylabel("Frequency")
plt.ylim([0,8000])
plt.legend()
plt.show()
```

```python
plt.hist(list(svhn_trainset.labels), color='r',label="TRAIN SET")
plt.hist(list(svhn_testset.labels),color='b', label="TEST SET")
plt.xlabel("Data Labels")
plt.ylabel("Frequency")
plt.title("Plot of SVHN TRAIN AND TRAIN DATASET")
plt.legend()
plt.show()
```

```python
plt.hist(list(svhn_extraset.labels))
plt.xlabel("Data Labels")
plt.ylabel("Frequency")
plt.title("Plot of SVHN EXTRA TRAIN SET")
plt.show()
```
